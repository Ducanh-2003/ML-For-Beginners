```python
import pandas as pd

pumpkins = pd.read_csv('../data/US-pumpkins.csv')
```

```python
pumpkins = pumpkins[pumpkins['Package'].str.contains('36 inch bins', case=True, regex=True)]
```

```python
pumpkins = pumpkins.loc[:,['Package', 'Low Price', 'High Price','Date']]
pumpkins.head()
price_mean = (pumpkins['Low Price'] + pumpkins['High Price']) / 2   
month = pd.to_datetime(pumpkins['Date']).dt.month
new = pd.DataFrame({'Month': month, 'Package': pumpkins['Package'], 'Low Price': pumpkins['Low Price'], 'High Price': pumpkins['High Price'], 'Price': price_mean})
```

```python
import matplotlib.pyplot as plt
price = new.Price
month = new.Month
plt.scatter(price, month, color='blue', s = 100)
plt.title('Matplotlib: Giá Bí ngô theo Tháng (36 inch bins)')
plt.xlabel('Giá trung bình ($)')
plt.ylabel('Tháng')
plt.grid(True, alpha = 0.3)
plt.show()
```

```python
import seaborn as sns
sns.scatterplot(data=new, x=price, y=month, s=100)

plt.title('Seaborn-scatter: Giá theo Tháng')
plt.xlabel('Giá trung bình ($)')
plt.ylabel('Tháng')
plt.grid(True, alpha=0.3)
plt.show()
```

* Seaborn: dễ hơn, làm việc mượt mà với pandas, chỉ cần khai báo DataFrame và tên cột, thư viện tự động trích xuất dữ liệu. Cú pháp ngắn gọn trực quan, style mặc định đẹp hơn.
* Matplotlib: Yêu cầu tách rời biến trước khi đưa vào hàm vẽ, hoạt động độc lập với cấu trúc DataFrame, yêu cầu tùy chỉnh nhiều
